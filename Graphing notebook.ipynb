# Graphs of equations given a list of x_values.

```python
import matplotlib.pyplot as plt
import pandas as pd
################################# USER DEFINED SPACE ##########################
# Set your x_axis and initialise the empty y_axis.
x_axis = [-10,-4,2]
y_axis = []
# Iterate over x_axis.
####################################
for x in x_axis:
    # EQUATION IS ENTERED HERE.
    y = (0.6 * x) + 4.4
    # Append result to y_axis list.
    y_axis.append(y)
#####################################
# Set your x and y markers that appear at the end of the positive x and y axis lines.
xmarker = 'x '
ymarker = 'y '
# Change your labels
x_label = ''
y_label = ''
title = ''
# Choose an appropriate step between x and y ticks.
x_step = 1
y_step = 1
###############################################################################
# Custom X and Y ticks.
for i in range(round(max(x_axis))+1,round(max(x_axis))+x_step+1):
    if i % x_step == 0:
        pos_x_max = i
for i in range(round(min(x_axis))-1,round(min(x_axis))-x_step-1,-1):
    if i % x_step == 0:
        neg_x_max = i
for i in range(round(max(y_axis))+1,round(max(y_axis))+y_step+1):
    if i % y_step == 0:
        pos_y_max = i
for i in range(round(min(y_axis))-1,round(min(y_axis))-y_step-1,-1):
    if i % y_step == 0:
        neg_y_max = i
xticks = [i for i in range(neg_x_max,pos_x_max+1,x_step)]
yticks = [i for i in range(neg_y_max,pos_y_max+1,y_step)]
# Change figure size/
plt.figure(figsize=(10,8))
# Plot the graph.
plt.plot(x_axis,y_axis,marker='o',markersize=10)
# Change the margin.
#plt.margins(x=0.1,y=0.1)
#Alter your x,y labels and title.
plt.xlabel(xlabel=x_label,fontsize=10)
plt.ylabel(ylabel=y_label,fontsize=10)
plt.title(label=title)
# Vertical and horizontal center lines.
plt.axhline(y = 0, color = 'black', linestyle = '-') 
plt.axvline(x = 0, color = 'black', linestyle = '-') 
# Annotate the end of the vertical and horizontal lines with your designated markers.
plt.text(
        # Xaxis
        xticks[-1]+x_step/10,
        # Yaxis
        y_step / -3,
        xmarker,
        fontsize=10)
plt.text(
        # Xaxis
        x_step / -10,
        # Yaxis
        yticks[-1]+y_step/10,
        ymarker,
        fontsize=10)
# Annotate the center lines with the x and y ticks and draw a grid.
for x in xticks:
    plt.text(x,x_step/10,x,rotation=0)
    if x != 0:
        plt.axvline(x = x, color = 'grey', linestyle = '--')
for y in yticks:
    plt.text(x_step/20,y,y,rotation=0)
    if y != 0:
        plt.axhline(y = y, color = 'grey', linestyle = '--') 
# Remove default x and y ticks.
plt.xticks([],fontsize=14)
plt.yticks([],fontsize=14)
# Create the table.
table = pd.DataFrame({
    'X Axis':x_axis,
    'Y Axis':y_axis
})
# Print the table.
print(table)
# Gradient or Slope calculation.
run = x_axis[1] - x_axis[0]
rise = y_axis[1] - y_axis[0]
gradient = rise / run
print('')
print('Gradient:',gradient)
# Display the graph.
plt.show()
```

# Drawing lines from their equations.

```python
import matplotlib.pyplot as plt
import pandas as pd
################################# USER DEFINED SPACE ##########################
# Set your x_axis and initialise the empty y_axis.
x_axis = [-0.5,-0.8]
y_axis = []
# Iterate over x_axis.
for x in x_axis:
    # EQUATION IS ENTERED HERE.
    y = (2.2 * x) +0.3
    y_equation = 'y = 2.2x + 0.3'
    # Append result to y_axis list.
    y_axis.append(round(y))
# Set your x and y markers that appear at the end of the positive x and y axis lines.
xmarker = 'x '
ymarker = 'y '
# Change your labels
x_label = ''
y_label = ''
title = ''
# Choose an appropriate step between x and y ticks.
x_step = 1
y_step = 1
###############################################################################
# Custom X and Y ticks.
for i in range(round(max(x_axis))+1,round(max(x_axis))+x_step+1):
    if i % x_step == 0:
        pos_x_max = i
for i in range(round(min(x_axis))-1,round(min(x_axis))-x_step-1,-1):
    if i % x_step == 0:
        neg_x_max = i
for i in range(round(max(y_axis))+1,round(max(y_axis))+y_step+1):
    if i % y_step == 0:
        pos_y_max = i
for i in range(round(min(y_axis))-1,round(min(y_axis))-y_step-1,-1):
    if i % y_step == 0:
        neg_y_max = i
xticks = [i for i in range(neg_x_max,pos_x_max+1,x_step)]
yticks = [i for i in range(neg_y_max,pos_y_max+1,y_step)]
# Change figure size/
plt.figure(figsize=(10,8))
# Plot the graph.
plt.plot(x_axis,y_axis,marker='o',markersize=10)
# Change the margin.
#plt.margins(x=0.1,y=0.1)
#Alter your x,y labels and title.
plt.xlabel(xlabel=x_label,fontsize=10)
plt.ylabel(ylabel=y_label,fontsize=10)
plt.title(label=title)
# Vertical and horizontal center lines.
plt.axhline(y = 0, color = 'black', linestyle = '-') 
plt.axvline(x = 0, color = 'black', linestyle = '-') 
# Add equation onto graph.
plt.text(
        # Xaxis
        xticks[1]+(x_step/2),
        # Yaxis
        yticks[1]+(y_step/2),
        y_equation,
        fontsize=15)
# Annotate the end of the vertical and horizontal lines with your designated markers.
plt.text(
        # Xaxis
        xticks[-1]+x_step/10,
        # Yaxis
        y_step / -3,
        xmarker,
        fontsize=10)
plt.text(
        # Xaxis
        x_step / -10,
        # Yaxis
        yticks[-1]+y_step/10,
        ymarker,
        fontsize=10)
# Annotate the center lines with the x and y ticks and draw a grid.
for x in xticks:
    plt.text(x,x_step/10,x,rotation=0)
    if x != 0:
        plt.axvline(x = x, color = 'grey', linestyle = '--')
for y in yticks:
    plt.text(x_step/20,y,y,rotation=0)
    if y != 0:
        plt.axhline(y = y, color = 'grey', linestyle = '--') 
# Remove default x and y ticks.
plt.xticks([],fontsize=14)
plt.yticks([],fontsize=14)
# Create the table.
table = pd.DataFrame({
    'X Axis':x_axis,
    'Y Axis':y_axis
})
# Print the table.
print(table)
# Gradient or Slope calculation.
run = x_axis[1] - x_axis[0]
rise = y_axis[1] - y_axis[0]
gradient = rise / run
print('')
print('Gradient:',gradient)
# Display the graph.
plt.show()
```
